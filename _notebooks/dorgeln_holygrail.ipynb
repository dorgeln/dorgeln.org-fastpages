## Intro
One of the things that's almost a given in the life of a SysAdmin is that the freshest code always comes with the worst documentation. Forcing you to choose between an older well documented version or choose to risk living on the cutting edge off github enjoying both the newest most exciting features and the hope of not being the first to run into a new bug.

Making any kind of system that promises to solve the various coding, testing and documentation system grievances the elusive holy grail of system administration that always seems just one extra tool away from being found.Most of the various DevOps tools are really helpful, but I still think it's time to try a a new approach for new and evolving areas like data sciences.

[Dorgeln.org](https://dorgeln.org/) is a testbed for this new approaches and you can give things a try for yourself by simply clicking the [Colab](https://colab.research.google.com) or [binder](https://mybinder.org) badge on top of every dorgeln.org post.

## Jupyter Notebooks

[JupyterLab](https://jupyter.org/) is a web based development environment for literate programming that allows you to mix programming and documentation at will. I'm using Jupyter for quite some while and it's by far my most favorite ways to test and learn any kind of new Python packages.

Which also makes keeping my Jupyter Notebooks alive for an extended period of time a constant struggle. Because by the time I end up running them again my system setup and Python configuration has almost certainly changed and I end up hitting problems I didn't have previously.

## fastpages

[Fastpages](https://github.com/fastai/fastpages) from  [fast.ai](https://www.fast.ai) is the new kid on the block. Jupyter always had the functionality to turn any Notebook into a website, but fastpages takes things a couple of steps further by automating everything with GitHub Actions and GitHub Pages. Plus the nice little feature of keeping [Altair](https://altair-viz.github.io) visualization fully interactive on a statically generated web site hosted by [GitHub](https://github.com) for free.

## binder
[Binder](https://mybinder.org/) has the lofty goal of making every Jupyter notebook reproducible and accessible by everyone from everywhere and pretty much deliver on that goal. Startup times can be slow and there are some hiccups at times. But if your looking for a place to make your notebooks usable with one click binder offers a lot more possibilities (except GPU support) then the [Colab](https://colab.research.google.com) offerings from Google

## docker
[Docker](https://www.docker.com) has become the workhorse of any DevOps application. Running docker, especially docker compose can be a pain in the ass at times. But the concept of building layer based image on top of each other is a stroke of genius and is pretty much the only approach of quickly building lots of versioned system images and that actually works.

## nbgitpuller
[nbgitpuller](https://jupyterhub.github.io/nbgitpuller/) is a small extension for Jupyter that allows you to pull notebooks into a running Jupyter Lab instance via git. Which is extremely useful, because it allows you to generate a generic docker image for JupyterLab and then pull any notebook you like into the image at startup.


## fastnotes
[dorgeln/fastnotes](https://github.com/dorgeln/fastnotes) is my personal addition to the ever evolving Jupyter ecosystem. Jupyter does provide plenty of ready to use [Docker Images](https://jupyter-docker-stacks.readthedocs.io/en/latest/using/selecting.html#jupyter-datascience-notebook), but none of them really  matched my use cases. Either they are too limited to be useful or too big and bloated to be fun to use on binder.

Which set my on the hunt on building my own Jupyter docker image that works fine both with fastpages and binder and after using every trick possible including using Alpine Linux and multistage builds. I actually succeeded in building an image that's actually smaller than [jupyter/minimal-notebook](https://hub.docker.com/r/jupyter/minimal-notebook) despite being packet with 700MB of nodejs and python libraries. 

## orbiter
[dorgeln/orbiter](https://github.com/dorgeln/orbiter) is the last little piece of the Dorgeln puzzle and it's almost trivial. It's a repository with an extra branch for every release of dorgeln/fastnotes and a branch called stable. Each containing a simple one line Dockerfile.

```bash
FROM dorgeln/fastnotes:${VERSION}
```

So all you need to do to share your Jupyter notebooks quickly with everyone is to create a [nbgitpuller link](https://jupyterhub.github.io/nbgitpuller/link.html?tab=binder) and use **https://github.com/dorgeln/orbiter** and branch __stable__ as your Git Environment Repository URL.  You can also choose a specific version as your branch and your notebook is assured to run without problem in the future to come.
